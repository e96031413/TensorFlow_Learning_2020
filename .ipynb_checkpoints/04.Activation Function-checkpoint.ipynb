```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+(np.e**(-x)))

x = np.arange(-6, 6, 0.1)

plt.plot(x, sigmoid(x))
plt.title('sigmoid')
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)
x = np.arange(-6, 6, 0.1)

plt.plot(x ,relu(x))
plt.title('relu')
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)
x = np.arange(-6, 6, 0.1)

plt.plot(x, tanh(x))
plt.title('tanh')
plt.show()
```

```python
def tanh(x):
    return np.sinh(x)/np.cosh(x)
x = np.arange(-6, 6, 0.1)

plt.plot(x, tanh(x))
plt.title('tanh')
plt.show()
```

```python
def softmax(x):
    return np.exp(x)/sum(np.exp(x))

x = np.array([1, 2, 3, 4, 1, 2, 3])

y = softmax(x)
print(y)
```

```python
def MSE(y, t):
    return 0.5*np.sum((y-t)**2)
```
